```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import train_test_split
%matplotlib inline

from graphviz import Source
from IPython.display import SVG, display, HTML
style = "<style>svg{width: 70% !important; height: 60% !important;} </style>" 
```

```python
iris_data = pd.read_csv('train_iris.csv')
test_iris_data = pd.read_csv('test_iris.csv')
```

```python
X_train = iris_data.drop(['species'], axis=1)
y_train = iris_data.species
X_test = test_iris_data.drop(['species'], axis=1)
y_test = test_iris_data.species
```

```python
np.random.seed(0)
```

```python
max_deprh_valuses = range(1, 100)
```

```python
scores_data = pd.DataFrame()
```

```python
for max_depth in max_deprh_valuses:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
#     mean_cross_val_score = cross_val_score(clf, X_train, y_train, cv=5).mean()
    
    temp_score_data = pd.DataFrame({'max_depth': [max_depth],
                                    'train_score': [train_score],
                                    'test_score': [test_score]})
    
    scores_data = pd.concat([scores_data, temp_score_data])
```

```python
scores_data
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], 
                           value_vars=['train_score', 'test_score'],
                          var_name='set_type', value_name='score')
```

```python
scores_data_long
```

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long)
```

```python
dog_and_cat = pd.read_csv('dogs_n_cats.csv')
```

```python
X_train = dog_and_cat.drop(['Вид'], axis=1)
y_train = dog_and_cat['Вид']
```

```python
dnc_clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5)
```

```python
dnc_clf.fit(X_train, y_train)
```

```python
test = pd.read_json('dataset_209691_15.txt')
```

```python
test
```

```python
results = dnc_clf.predict(test)
```
