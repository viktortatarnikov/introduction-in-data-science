```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
df = pd.read_csv('train_data_tree.csv')
```

```python
df
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
X = df.drop('num', axis=1)
y = df.num
```

```python
clf.fit(X, y)
```

```python
tree.plot_tree(clf, feature_names=list(X),
               class_names=['0', '1'],
               filled=True);
```
